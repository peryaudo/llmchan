```python
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from glob import glob
import re
```

```python
lines = [line.split('<>') for line in open("scraped/1495000082.dat").readlines()]
title = lines[0][4]
comments = [BeautifulSoup(line[3]).get_text(separator = '\n', strip = True) for line in lines]
```

```python
for c in [c for c in comments if len(c) < 97 and ">>" not in c and "ttp" not in c][200:300]:
    print(c)
    print("========================================")
```

```python
titles = [re.sub(r'★.+$', '', title.replace('[無断転載禁止]', '').replace('©2ch.net', '').strip()) for title in titles]
titles = [title for title in titles if '批判要望' not in title]
```

```python
for title in titles[:200]:
    print(title)
```

### Conclusion

* Use older data to avoid spams
* Filter comments longer than 100, they tend to be non-comments/irrelevant
* For the same reason, exclude comments including 'ttp'
* The majority of comments look relevant, perhaps no easy way to quality gate
* The sizable portion of comments contain '>>', perhaps OK to exclude them for now but they might be potential source of high quality comments
